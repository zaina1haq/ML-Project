#### 2.6 Mapping featuers Using domain knowledge

### Grouping Diagnosis Codes into Clinical Categories

Diagnosis codes were grouped into clinically meaningful disease categories such as diabetes, cardiovascular, respiratory, and digestive conditions. This grouping reduces the high dimensionality of raw ICD-9 codes while preserving the medical relevance of the diagnoses.

By organizing diseases into broader clinical categories, the model can better capture underlying health patterns and disease severity, leading to more stable learning and improved prediction of hospital readmission risk.


````python
def icd9_group(diag):
    if pd.isna(diag):
        return "Unknown"

    diag = str(diag).strip()

    # External causes
    if diag.startswith("E"):
        return "Injury"

    # V codes (history / supplementary)
    if diag.startswith("V"):
        return "Other"

    try:
        code = int(float(diag.split('.')[0]))
    except:
        return "Unknown"

    if code == 250:
        return "Diabetes"

    elif 390 <= code <= 459:
        return "Circulatory"

    elif 460 <= code <= 519:
        return "Respiratory"

    elif 520 <= code <= 579:
        return "Digestive"

    elif 580 <= code <= 629:
        return "Genitourinary"

    elif 710 <= code <= 739:
        return "Musculoskeletal"

    elif 140 <= code <= 239:
        return "Neoplasms"

    else:
        return "Other"

````

````python
for col in ["diag_1", "diag_2", "diag_3"]:
    data[col] = data[col].apply(icd9_group)
data['diag_1'].value_counts()
````

### Medication Change Count

A new feature **med_change_count** is created by counting how many medications have values **'Up'** or **'Down'** in each row.  
This represents the number of medication changes per patient encounter.


````python
import pandas as pd

# List of medication columns
medication_columns = [
    'metformin','repaglinide','nateglinide','chlorpropamide',
    'glimepiride','acetohexamide','glipizide','glyburide',
    'tolbutamide','pioglitazone','rosiglitazone','acarbose',
    'miglitol','troglitazone','tolazamide',
    'glyburide-metformin','glipizide-metformin',
    'glimepiride-pioglitazone','metformin-rosiglitazone',
    'metformin-pioglitazone','insulin'
]

# Function to count medication changes (Up or Down) per row
def count_med_changes(row):
    return row[medication_columns].isin(['Up', 'Down']).sum()

# Create new feature: number of medication changes
data['med_change_count'] = data.apply(count_med_changes, axis=1)

# Check med_change_count by change column (validation)
print(data.groupby('change')['med_change_count'].describe())

# Drop redundant column
data.drop(columns="change", inplace=True)

# Preview results
print(data[['med_change_count']].head(10))
````

### Combination Drug Simplification

For combination drugs, the values **Steady**, **Up**, and **Down** were merged into a single category called **Taken**, while patients who did not take the drug remain labeled as **No**.

This simplifies the feature, reduces noise from rare cases, and still indicates whether the patient used the medication.


````python
# Merge Steady, Up, and Down into one category: 'Taken'
data['glyburide-metformin'] = data['glyburide-metformin'].replace({
    'Steady': 'Taken',
    'Up': 'Taken',
    'Down': 'Taken'
})

# Simplify other combination drugs to 'Taken'
data['glipizide-metformin'] = data['glipizide-metformin'].replace({
    'Steady': 'Taken'
})

data['glimepiride-pioglitazone'] = data['glimepiride-pioglitazone'].replace({
    'Steady': 'Taken'
})

data['metformin-rosiglitazone'] = data['metformin-rosiglitazone'].replace({
    'Steady': 'Taken'
})

data['metformin-pioglitazone'] = data['metformin-pioglitazone'].replace({
    'Steady': 'Taken'
})

# Check distribution after merging
print(data['glyburide-metformin'].value_counts())
````

### Feature Engineering of Medicines

A new feature **treatment_complexity** is created to represent the complexity of diabetes treatment.

- **0**: Metformin only  
- **1**: Oral medications only  
- **2**: Combination drugs  
- **3**: Insulin treatment  

This feature summarizes medication usage into a single, meaningful indicator that reflects treatment intensity.


````python
# List of oral diabetes drugs
oral_drugs = [
    'metformin','repaglinide','nateglinide','chlorpropamide',
    'glimepiride','acetohexamide','glipizide','glyburide',
    'tolbutamide','pioglitazone','rosiglitazone','acarbose',
    'miglitol','troglitazone','tolazamide'
]

# List of combination drugs
combination_drugs = [
    'glyburide-metformin','glipizide-metformin',
    'glimepiride-pioglitazone','metformin-rosiglitazone',
    'metformin-pioglitazone'
]

# Default: oral medications
data['treatment_complexity'] = 1

# Assign treatment complexity for each patient
for i, row in data.iterrows():

    # Insulin treatment
    if row['insulin'] != 'No':
        data.at[i, 'treatment_complexity'] = 3

    # Combination drugs
    elif any(row[drug] != 'No' for drug in combination_drugs):
        data.at[i, 'treatment_complexity'] = 2

    else:
        # Count oral drugs used
        oral_count = 0
        for drug in oral_drugs:
            if row[drug] != 'No':
                oral_count += 1

        # Metformin only
        if oral_count == 1 and row['metformin'] != 'No':
            data.at[i, 'treatment_complexity'] = 0

# Check results
print(data['treatment_complexity'].value_counts())
````

- **Recall** is used as the scoring metric to focus on correctly identifying readmitted patients.
- Cross-validation is applied to select the best-performing model.

The best model and its optimal parameters are then extracted for evaluation.


````python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Parameters to try
param_grid = {
    'n_estimators': [300, 500],        # Number of trees
    'max_depth': [5, 7, 9],            # Tree depth
    'min_samples_leaf': [3, 5],        # Min samples per leaf
    'min_samples_split': [5],          # Min samples to split
    'class_weight': [{0:1, 1:3}, {0:1, 1:4}],  # Handle imbalance
    'criterion': ['gini', 'entropy']   # Split metric
}

# Base Random Forest model
rf_base = RandomForestClassifier(
    random_state=42,
    n_jobs=-1                          
)

# Grid search using recall
grid = GridSearchCV(
    estimator=rf_base,
    param_grid=param_grid,
    scoring='recall',                  # Focus on recall
    cv=3,
    verbose=2
)

# Train and search best model
grid.fit(X_train, y_train)

# Get best model
best_rf = grid.best_estimator_

# Show best parameters
print("Best Parameters:", grid.best_params_)
````

````python
from sklearn.metrics import classification_report, roc_auc_score

y_prob = best_rf.predict_proba(X_test)[:, 1]
````

### Threshold Evaluation

Multiple thresholds are tested to evaluate model performance and select the best one.


````python
# Import required libraries
from sklearn.metrics import classification_report
from IPython.display import display, Markdown

# Different thresholds to test
thresholds = [0.5, 0.4, 0.3, 0.25, 0.2, 0.15]

# Save results here
output = ""

# Evaluate model at each threshold
for t in thresholds:
    # Convert probabilities to predictions
    y_pred = (y_prob >= t).astype(int)

    # Store results
    output += "\n" + "=" * 50 + "\n"
    output += f"Threshold = {t}\n"
    output += classification_report(y_test, y_pred)

    # Show results
    display(Markdown(f"### Threshold = {t}"))
    display(Markdown(f"```\n{classification_report(y_test, y_pred)}\n```"))

# Write all results to a file
with open("ALL_RESULTS.txt", "w", encoding="utf-8") as f:
    f.write(output)
````

### Final Model Evaluation

The model is evaluated using a final threshold of **0.3**.

Performance is assessed with:
- **Classification report**
- **ROC AUC**
- **Balanced accuracy**

This provides a fair evaluation of the model, especially with imbalanced classes.


````python
from sklearn.metrics import (
    classification_report,
    roc_auc_score,
    balanced_accuracy_score
)

# Set the final decision threshold
final_threshold = 0.3

# Get predicted probabilities for the positive class
y_prob = best_rf.predict_proba(X_test)[:, 1]

# Convert probabilities to final binary predictions
y_pred_final = (y_prob >= final_threshold).astype(int)

# Show classification results
print("Final Evaluation (Threshold = 0.3)")
print(classification_report(y_test, y_pred_final))

# Calculate ROC AUC score
print("ROC AUC:", roc_auc_score(y_test, y_prob))

# Calculate balanced accuracy (good for imbalanced data)
print("Balanced Accuracy:", balanced_accuracy_score(y_test, y_pred_final))
````
